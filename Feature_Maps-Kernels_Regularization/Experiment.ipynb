```python
import numpy as np
import matplotlib.pyplot as plt
```

Generating Data

$$y_{\text{true}} = \log(X^2) \cdot coef_{\text{true}} + \text{offset}
     $$


```python
# Generating the data
n_samples = 1000
d_samples = 3
np.random.seed(42)

# X
X = np.random.random((n_samples, d_samples)) * 2 - 1
# y_true
coeff_true = [-10, -15, -7.5]
offset = 2
y_true = np.log(X**2) @ coeff_true + offset
# y_noisy
gaussian_noise = np.random.normal(loc=0, scale=10, size=(1000))
y_noisy = y_true + gaussian_noise
# train/test
split = int(n_samples / 10 * 8)
X_train, X_test = X[0:split, :], X[split:, :]
y_train, y_test = y_noisy[0:split], y_noisy[split:]
```

Plot data

```python
# Plotting the scatter plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i in range(d_samples):
    axes[i].scatter(X[:, i], y_noisy, alpha=0.5)
    axes[i].set_xlabel(f'Feature {i+1}')
    axes[i].set_ylabel('Noisy Label')
    axes[i].set_title(f'Feature {i+1} vs Noisy Label')

plt.tight_layout()
plt.show()
```
