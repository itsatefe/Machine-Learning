```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Q1

```python
dataset = pd.read_csv("adult.csv")
```

# Q2

```python
num_samples, num_features = dataset.shape
print("number of samples: ", num_samples)
print("number of features: ", num_features)
```

```python
np.unique(dataset['income'])
```

# Q3

```python
columns = ['workclass', 'education', 'marital.status', 'race','sex', 'income']
dataset = dataset[columns]
```

```python
num_duplicates_before = len(dataset)
dataset.drop_duplicates(inplace=True)
num_duplicates_after = len(dataset)
num_duplicates_removed = num_duplicates_before - num_duplicates_after

print("Number of duplicates before dropping:", num_duplicates_before)
print("Number of duplicates after dropping:", num_duplicates_after)
print("Number of duplicates removed:", num_duplicates_removed)
```

# Q4

```python
dataset.replace('?', pd.NA, inplace=True)
dataset.dropna(inplace=True)
nan_values_remaining = dataset.isna().sum().sum()
print("Number of NaN values remaining:", nan_values_remaining)
```

# Q5

```python
print("number of remaining samples: ", len(dataset))
```

```python

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 12))

for i, column in enumerate(dataset.columns):
    row = i // 3
    col = i % 3
    sns.countplot(data=dataset, x=column, ax=axes[row][col])
    axes[row][col].set_title('Histogram of ' + column)
    axes[row][col].set_xlabel('Categories')
    axes[row][col].set_ylabel('Frequency')
    axes[row][col].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
